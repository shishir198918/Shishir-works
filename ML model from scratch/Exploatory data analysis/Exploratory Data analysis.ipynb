```python
import matplotlib.pyplot as plt
import numpy as np

# Example data
x = np.linspace(0, 2 * np.pi, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# Using a figure with subplots
fig, axs = plt.subplots(2, 1, figsize=(8, 6))

# Subplot 1 (top)
axs[0].plot(x, y1, label='sin(x)')
axs[0].set_title('Sin(x)')

# Subplot 2 (bottom)
axs[1].plot(x, y2, label='cos(x)', color='orange')
axs[1].set_title('Cos(x)')

# Adjust layout for better spacing
plt.tight_layout()

# Show the plot
plt.show()
```

```python
d=d1.copy()
d["RestingBP"]
```

```python


## sns.displot() used for showing distribution of that data like normal, skew, exponential,e.t.c
sns.distplot(d["MaxHR"])
sns.displot(d["Cholesterol"])
sns.distplot(d["RestingBP"])



```

```python
mean_MaxHR=d["MaxHR"].mean()
std_MaxHR=d["MaxHR"].std()
mean_Cholesterol=d["Cholesterol"].mean()
std_Cholesterol=d["Cholesterol"].std()
```

```python
mean_MaxHR,std_MaxHR
```

```python
## Standarization
d["Z_MaxHR"]=((d["MaxHR"]-mean_MaxHR)/std_MaxHR)
d["Cholesterol"]=((d["Cholesterol"]-mean_Cholesterol)/std_Cholesterol)
```

```python
sns.distplot(d["Z_MaxHR"])
sns.distplot(d["Cholesterol"])
```

#Using 3 and -3 logic to capping some values 
def capping(list1,cap1,cap0):
    list2=[]
    for i in list1:
        if i>cap1:
            i=cap1
            list2.append(i)
        elif i<cap0:
            i=cap0
            list2.append(i)
        else:
            list2.append(i)
            
    return list2   

```python
dictonary={"rank":[7,6,9,8,11,16,20,100,1,5,8,3,4]}
dic1=pd.DataFrame(dictonary)
dic1["rank"]=capping(dic1["rank"],15,5)
```

Interquartile Range [IQR]
The data range that contains the middle 50%
of the entries.

IQR = Q3 - Q1

The values which are below Q1 - 1.5 IQR
or above Q3 + 1.5 IQR are outliers.

This is called the 1.5 x IQR rule.
