```python
#Retriving the list of punctuation 
import string
# getting the list of puncation
ignore=string.punctuation
ignore=ignore.replace(".","")
# for more efficient code 
def remove_punc1(text):
    return text.translate(str.maketrans("","",ignore))
 
#stopwords
from nltk.corpus import stopwords
stop_words = set(stopwords.words('english'))
def removal_lowering_stopword(text):
    l=[]
    text_list=text.lower().split()
    for w in text_list:
        if w  not in stop_words:
            l.append(w)
    # convert into string again         
    text=" ".join(l)  
    return text


def read_n_combine(path):
    final_list=[]
    all_text=""
    for file in os.listdir(path):
        file_path = os.path.join(path, file)
        if os.path.isfile(file_path) and file.endswith(".txt"):
            file= open(file_path, 'r')
            # Storing into a string 
            text=file.read()
            # remove punctuation
            text1=remove_punc1(text)
            # lowering and removing stop word 
            text2=removal_lowering_stopword(text1)
            all_text=all_text + text2 + "\n"
            file.close()
    output_path=os.path.join(path,"all_part.txt")        
    all_part=open(output_path,"w")
    all_part.write(all_text)
    all_part.close()
    return( output_path)
```

```python
story_path=read_n_combine("book")
story=open(story_path,"r")
story_text=story.read()
sentence_token=sentence_tokenizer(story_text)
word_token=word_tokenizer(story_text)
story.close()
```
