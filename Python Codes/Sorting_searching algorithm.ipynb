## Searching in a sorted list 


```python
def binary_search(v,l):
    if l==[]:
        return False
    m=len(l)//2
    if l[m]==v:
        return True 
    elif v<l[m]:
        return(binary_search(v,l[:m]))
    elif v>l[m]:
        return (binary_search(v,l[m:]))
```

```python
binary_search(7,[2,4,6,7,9,10])
```

```python
# insertion sort 
def insert(l,x):
    index=0
    for i in range(len(l)):
        if l[i]>x:
            index=i
            return l[:index]+[x]+l[index:]   
    return l+[x]    
       

    
def isort(L):
    m=L[0]
    
    if len(L)<=1:
        return L
    else:
        L.remove(m)
        
        return insert(isort(L),m) 
```

```python
# collage code insertion sort 

def insertion(l):
    n=len(l)
    if n<1:
        return l
    
    for i in range(len(l)):
        j=i
        while(j>0 and l[j]<l[j-1]):
            l[j],l[j-1]=l[j-1],l[j]
            j=j-1
            
    return l        
    
    

```

## Merge sort 



```python
def mymerge(A,B):# assuming A,B is sorted 
    i,j=0,0
    C=[]
    while(i<len(A) and j<len(B)):
        if A[i]<B[j]:
            C.append(A[i])
            i=i+1
        else:
            C.append(B[j])
            j=j+1
            
    if i==len(A):
        C.extend(B[j:])
    else:
        C.extend(A[i:])
    return C  


def mergesort(A):
    if len(A)<=1:
        return A
    L=mergesort(A[(len(A)//2):])
    R=mergesort(A[:(len(A)//2)])
    B=mymerge(L,R)
    return(B)
```

```python
mergesort([2,1,2,4,5,6,1,4,5])
```
