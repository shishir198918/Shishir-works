```python
import time
class Timer:
    def __init__(self):
        self._start_time=0
        self._elapsed_time=0
    def start(self):
        self._start_time=time.perf_counter()
        
    def stop(self):
        self._elapsed_time= time.perf_counter()-(self._start_time)
    
    def elapsed(self):
        return (self._elapsed_time)
    
t1=Timer()
n=999
t1.start()
while(n>0):
    n=n-1
    
t1.stop()
p=t1.elapsed()
print(p)
```

Twin primes are pairs of prime numbers that differ by 2. For example (3, 5), (5, 7), and (11,13) are twin primes.
Write a function Twin_Primes(n, m) where n and m are positive integers and n < m , that returns all unique twin primes between m and n (both inclusive). The function returns a list of tuples and each tuple (a,b) represents one unique twin prime where n <= a < b <= m.

```python
def isprime(n):
    l=[]
    for i in range(1,n+1):
        if n%i==0:
            l.append(i)
            
    if len(l)==2:
        return n
    else:
        return 0
        
def listprime(n,m):
    l1=[]
    for i in range(n,m+1):
        if isprime(i)==0:
            pass
        else:
            l1.append(isprime(i))
    return l1

      

def pairprime(n, m):
    l2 = listprime(n, m)
    l3 = []
    for i in range(1, len(l2)):
        if l2[i] - l2[i - 1] == 2:
            a, b = l2[i - 1], l2[i]
            l3.append((a, b))
    return l3

        
        
```

```python
#Q-2

class Triangle:
    def __init__(self,a,b,c):
        self.a=a
        self.b=b
        self.c=c
    def Is_valid(self):
        if (self.a+self.b>self.c)&(self.b+self.c>self.a)&(self.a+self.c>self.b):
            return "valid"
        else:
            return "Invalid"
    def Side_Classification(self):
        case=self.Is_valid()
        if case =="valid":
            if self.a==self.b==self.c:
                return "Equilateral"
            elif (self.a==self.b)or(self.b==self.c)or(self.a==self.c):
                return "Isosceles"
            else:
                return "Scalene"
        else:
            return self.Is_valid()
        
    def Angle_Classification(self):
        case=self.Is_valid()
        if case=="valid":
            a,b,c=self.a**2,b.self**2,self.c**2
            if a+b>c:
                return "Acute"
            elif a+b==c:
                return "Right"
            else:
                return "Obtuse"
    
    def Area(self):
        case=self.Is_valid()
        s=(self.a+self.b+self.c)/2
        if case=="valid":
            return (s*(s-self.a)*(s-self.b)*(s-self.c))**.5
        
```

```python
T=Triangle(3,4,5)

T.Side_Classification()
T.Area()
```
